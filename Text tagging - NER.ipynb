# Named Entity Recognition(NER)

NER is powerful tool for turning unstructured text into structured information

```python
import spacy
```

displacy = built-in visualisation tool (highlights entities)

tokenizer = breaks text to tokens

html to display from IPython - to render information from notebook

```python
from spacy import displacy
from spacy import tokenizer
from IPython.display import HTML,display
import re
```

```python
google_text = "Google was founded on September 4, 1998, by computer scientists Larry Page and Sergey Brin while they were PhD students at Stanford University in California. Together they own about 14% of its publicly listed shares and control 56% of its stockholder voting power through super-voting stock. The company went public via an initial public offering (IPO) in 2004. In 2015, Google was reorganized as a wholly owned subsidiary of Alphabet Inc. Google is Alphabet's largest subsidiary and is a holding company for Alphabet's internet properties and interests. Sundar Pichai was appointed CEO of Google on October 24, 2015, replacing Larry Page, who became the CEO of Alphabet. On December 3, 2019, Pichai also became the CEO of Alphabet."

print(google_text)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(google_text)
```

=========================================================

in spacy -  label: a numeric ID of the entitiy

        label_ : a human-readable label
Some common entity labels:

PERSON → people

ORG → organizations

GPE → countries, cities, states

DATE → dates

TIME → times

MONEY → money amounts

LOC → locations

PRODUCT → products

.===================================================

```python
for word in spacy_doc.ents:
     print(word.text, word.label_)
```

## ===============================================

displacy.render(...) takes a spaCy Doc and turns it into HTML markup.

style='ent'
This tells spaCy to render entities.
So it will highlight things like:
* person names
* organizations

jupyter=False
This tells displacy.render to return the HTML as a string instead of trying to display it automatically inside Jupyter.

output -> <span class="entity">Google<...>ORG</span>

display(HTML()) - This takes that raw HTML string and tells Jupyter/IPython:
* interpret this as actual HTML
* render it visually in the notebook

## ===================================================


```python
html = displacy.render(spacy_doc, style = 'ent', jupyter = False)
display(HTML(html))
```

## how it looks after cleaning 


```python
google_text_clean = re.sub(r'[^\w\s]','',google_text).lower()
```

```python
google_text_clean
```

```python
spacy_doc_clean = nlp(google_text_clean)
```

```python
for word in spacy_doc_clean.ents:
     print(word.text, word.label_)
```

```python
html = displacy.render(spacy_doc_clean, style = 'ent', jupyter = False)
display(HTML(html))
```
